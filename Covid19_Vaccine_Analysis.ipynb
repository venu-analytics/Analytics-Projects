## Covid 19 Vacchination Analysis with Python

This dataset is taken from Kaggle https://www.kaggle.com/gpreda/covid-world-vaccination-progress

The following are the respected python packages for the analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

Now let's import the data and explore the data before we start analysing it

```python
data = pd.read_csv("/content/country_vaccinations.csv")
data.head()
```

Now, check the type of vaccines available in this dataset 

```python
data.vaccines.value_counts()
```

##### Now, Let's explore which vaccine is taken by which country by creating another dataframe from major dataframe

```python
df = data[['country','vaccines']]
df.head()
```

Now, Let's check each vaccine is taken by how many countries

```python
dict_ = {}
for i in df.vaccines.unique():
  dict_[i] = [df.country[j] for j in df[df.vaccines==i].index]

vaccines = {}
for key, value in dict_.items():
  vaccines[key] = set(value)
for key, value in vaccines.items():
  print(f"{key}:{value}")
```

Now, let's visualize the data to check better understanding of combination of vaccines every country is using.

```python
import plotly.express as px
import plotly.offline as py

vaccine_map = px.choropleth(data, locations = 'iso_code', color = 'vaccines')
vaccine_map.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
vaccine_map.show()
```
